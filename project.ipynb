#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
get_trf = ds[(ds['level5'] == 'Getafe')]
alc_trf = ds[(ds['level5'] == 'Alcorcón')]

price_m2_get = ((get_trf['price'].mean()) / (get_trf['surface'].mean()))
price_m2_alc = ((alc_trf['price'].mean()) / (alc_trf['surface'].mean()))

print(f'The price per square meter in Getafe is {price_m2_get}, whilst in Alcorcon is {price_m2_alc}')

# We can say is cheaper to buy in Getafe than Alcorcon
```

#### Reload the data, is the average of "Valdemorillo" and "Galapagar" prices the same?

```python
val_trf = ds[(ds['level5'] == 'Valdemorillo')]
gal_trf = ds[(ds['level5'] == 'Galapagar')]

price_avg_val = val_trf['price'].mean()
price_avg_gal = gal_trf['price'].mean()

print(f'Average price Valdemorillo: {price_avg_val}\nAverage price Galapagar: {price_avg_gal}')

# They differ in about 3800 money units, leading Valdemorillo
```

#### What happens if we do the same analysis for price/m2 (★★☆) 

```python
val_trf = ds[(ds['level5'] == 'Valdemorillo')]
gal_trf = ds[(ds['level5'] == 'Galapagar')]

price_m2_val = ((val_trf['price'].mean()) / (val_trf['surface'].mean()))
price_m2_gal = ((gal_trf['price'].mean()) / (gal_trf['surface'].mean()))

print(f'The price per square meter in Valdemorillo is {price_m2_val}, whilst in Galapagar is {price_m2_gal}')

# While Valdemorillo has higher mean price than Galapagar, in this last place the prices are higher when considered the surface, meaning that houses in Valdemorillo are relatively bigger.
```
